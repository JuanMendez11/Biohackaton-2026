```python
from rdkit import Chem
from rdkit.Chem import Descriptors

# 1. Definimos la función detallada
def analizar_heteroatomos(smiles, nombre="Molecula"):
    mol = Chem.MolFromSmiles(smiles)
    if mol is None: 
        return "Error con el SMILES"
    
    heteroatomos_encontrados = []
    
    for atomo in mol.GetAtoms():
        # Si el número atómico NO es 6 (C) ni 1 (H), es un heteroátomo
        if atomo.GetAtomicNum() not in (1, 6):
            simbolo = atomo.GetSymbol()
            heteroatomos_encontrados.append(simbolo)
            
    cantidad = len(heteroatomos_encontrados)
    
    print(f"--- Análisis Detallado: {nombre} ---")
    print(f"Total de heteroátomos: {cantidad}")
    print(f"Átomos exactos encontrados: {heteroatomos_encontrados}\n")

# ==========================================
# Ejecución con la Aspirina
# ==========================================

# SMILES de la Aspirina
smiles_aspirina = "CC(=O)Oc1ccccc1C(=O)O"

# Probamos el Método 1 (Nuestra función iterativa)
analizar_heteroatomos(smiles_aspirina, "Aspirina")

# Probamos el Método 2 (El One-Liner directo de RDKit para tu dataset)
mol_aspirina = Chem.MolFromSmiles(smiles_aspirina)
total_rapido = Descriptors.NumHeteroatoms(mol_aspirina)

print(f"--- Análisis Rápido (One-Liner) ---")
print(f"Total directo extraído para el dataset: {total_rapido}")
```

```python
import pandas as pd
from rdkit import Chem
from rdkit import RDLogger 

# Silenciamos los errores de C++ en la consola
lg = RDLogger.logger()
lg.setLevel(RDLogger.CRITICAL)

# Precompilamos el patrón SMARTS fuera de la función para que sea ultrarrápido
PATRON_HETEROAROMATICO = Chem.MolFromSmarts("[a;!#6]")

def es_cromoforo(smiles):
    """
    Evalúa si una molécula tiene al menos un anillo heteroaromático
    usando coincidencia de subestructuras (A prueba de versiones RDKit).
    """
    try:
        mol = Chem.MolFromSmiles(smiles)
        
        if mol is None:
            return False
            
        # HasSubstructMatch devolverá True si encuentra el patrón
        return mol.HasSubstructMatch(PATRON_HETEROAROMATICO)
        
    except Exception as e:
        print(f"Error procesando {smiles}: {e}")
        return False
```

```python
import pubchempy as pcp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from rdkit import Chem
from rdkit import RDLogger

# ==========================================
# 1. PREPARACIÓN Y FUNCIONES CORE
# ==========================================

# Silenciamos RDKit para que no ensucie la consola
lg = RDLogger.logger()
lg.setLevel(RDLogger.CRITICAL)

# Precompilamos el patrón SMARTS (Cualquier átomo aromático que NO sea Carbono)
PATRON_HETEROAROMATICO = Chem.MolFromSmarts("[a;!#6]")

def es_cromoforo(smiles):
    """Evalúa si la molécula tiene un anillo heteroaromático."""
    try:
        # Aseguramos que haya texto
        if not smiles or not isinstance(smiles, str): 
            return False
            
        mol = Chem.MolFromSmiles(smiles)
        if mol is None: 
            return False
            
        return mol.HasSubstructMatch(PATRON_HETEROAROMATICO)
    except:
        return False

# ==========================================
# 2. DESCARGA MASIVA OPTIMIZADA
# ==========================================

start_cid = 3000
count = 150  # Descargamos 150 para asegurar una buena muestra
cids = list(range(start_cid, start_cid + count))

print(f"Descargando {count} compuestos de PubChem (Esto toma ~2 segundos)...")
# Al pasar la lista completa, PubChemPy hace UNA sola llamada a la API. ¡Es rapidísimo!
compuestos = pcp.get_compounds(cids, 'cid')

data_list = []
for c in compuestos:
    # Usamos .smiles para evitar el deprecation warning
    smiles = c.smiles 
    if smiles:
        data_list.append({
            'CID': c.cid,
            'Nombre': c.iupac_name if c.iupac_name else f"CID_{c.cid}",
            'Peso_Molecular': c.molecular_weight,
            'LogP': c.xlogp,
            'SMILES': smiles
        })

df = pd.DataFrame(data_list)
# Limpiamos moléculas que no tengan datos numéricos para evitar errores al graficar
df = df.dropna(subset=['Peso_Molecular', 'LogP']) 

# ==========================================
# 3. ANÁLISIS: APLICAR NUESTRA FUNCIÓN
# ==========================================

print("Aplicando filtro estructural de Cromóforos a las moléculas...")
df['Es_Cromoforo'] = df['SMILES'].apply(es_cromoforo)

# Imprimir estadísticas rápidas
conteo = df['Es_Cromoforo'].value_counts()
print(f"Dataset final: {len(df)} moléculas procesadas con éxito.")
print(f"--> Cromóforos detectados: {conteo.get(True, 0)}")
print(f"--> No Cromóforos (Moléculas comunes): {conteo.get(False, 0)}")

# ==========================================
# 4. VISUALIZACIÓN
# ==========================================

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Configuramos el gráfico para que la diferencia sea obvia
sns.scatterplot(
    data=df, 
    x='Peso_Molecular', 
    y='LogP', 
    hue='Es_Cromoforo', 
    style='Es_Cromoforo',
    palette={True: '#FF3333', False: '#3388FF'}, # Rojo vibrante para pigmentos, azul apagado para el resto
    markers={True: 'o', False: 'X'}, # Círculos vs Cruces
    s=120, # Tamaño de los puntos
    alpha=0.8
)

plt.title('Distribución Espacial: Cromóforos vs Compuestos Comunes', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Peso Molecular (g/mol) -> Tamaño de la molécula')
plt.ylabel('Solubilidad (XLogP) -> Menos es más soluble en agua')
plt.legend(title='¿Es Cromóforo?', loc='upper right')

# Opcional: Mostrar los nombres de los 3 cromóforos más pesados para ver qué son
cromoforos = df[df['Es_Cromoforo'] == True].sort_values(by='Peso_Molecular', ascending=False).head(3)
for _, row in cromoforos.iterrows():
    plt.annotate(row['Nombre'][:15] + '...', # Acortamos el nombre si es muy largo
                 (row['Peso_Molecular'], row['LogP']),
                 xytext=(10, -5), textcoords='offset points', 
                 fontsize=8, color='#FF3333', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ==========================================
# 1. CARGAR TUS DATOS LOCALES
# ==========================================
ruta_csv = "data/PubChem_compound_reactive_dye.csv" # Cambia esto si tu archivo tiene otro nombre o está en otra carpeta

try:
    df_sinteticos = pd.read_csv(ruta_csv)
    
    # Si tus columnas se llaman distinto en el CSV, descomenta y usa esto para renombrarlas:
    # df_sinteticos.rename(columns={'Nombre_Mol': 'Nombre', 'MW': 'Peso_Molecular', 'XLogP': 'LogP'}, inplace=True)
    
    # Creamos la etiqueta para que tu dataset resalte en el gráfico
    df_sinteticos['Categoria'] = 'Sintético (CSV)'
    print(f"✅ Se cargaron {len(df_sinteticos)} compuestos sintéticos de tu CSV.")

except FileNotFoundError:
    print(f"⚠️ Error: No se encontró el archivo '{ruta_csv}'.")
    df_sinteticos = pd.DataFrame() # Creamos uno vacío para que el código no explote

# ==========================================
# 2. PREPARAR EL DATASET ANTERIOR Y FUSIONAR
# ==========================================
if not df_sinteticos.empty:
    # Transformamos tu columna True/False anterior en etiquetas de texto
    # (Asumiendo que el DataFrame del paso anterior todavía se llama 'df')
    df['Categoria'] = df['Es_Cromoforo'].map({
        True: 'Cromóforo (PubChem)',
        False: 'Común (PubChem)'
    })
    
    # Unimos ambos DataFrames poniendo uno debajo del otro
    # ignore_index=True resetea los números de fila para que no haya duplicados
    df_final = pd.concat([df, df_sinteticos], ignore_index=True)
    
    # ==========================================
    # 3. GRAFICAR LA TRIPLE COMPARATIVA
    # ==========================================
    plt.figure(figsize=(12, 8))
    sns.set_theme(style="whitegrid")
    
    # Definimos la paleta de colores (Tus sintéticos irán en un amarillo/naranja tóxico para destacar)
    colores = {
        'Cromóforo (PubChem)': '#FF3333',     # Rojo vibrante
        'Común (PubChem)': '#3388FF',         # Azul apagado
        'Sintético (CSV)': '#FFAA00'          # Naranja/Dorado
    }
    
    # Diferenciamos por formas geométricas
    marcadores = {
        'Cromóforo (PubChem)': 'o',           # Círculos
        'Común (PubChem)': 'X',               # Cruces
        'Sintético (CSV)': 's'                # Cuadrados (squares)
    }
    
    sns.scatterplot(
        data=df_final,
        x='Peso_Molecular',
        y='LogP',
        hue='Categoria',
        style='Categoria',
        palette=colores,
        markers=marcadores,
        s=150,           # Tamaño general de los puntos
        alpha=0.85,      # Un poco de transparencia para ver si se solapan
        edgecolor='black'# Borde negro para darles contraste
    )
    
    plt.title('Comparativa Espacial: PubChem vs Tintes Sintéticos', fontsize=16, fontweight='bold', pad=15)
    plt.xlabel('Peso Molecular (g/mol)', fontsize=12)
    plt.ylabel('Solubilidad (XLogP)', fontsize=12)
    
    # Movemos la leyenda afuera del gráfico para que no tape los datos
    plt.legend(title='Tipo de Molécula', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=11)
    
    plt.tight_layout()
    plt.show()
```
